```python
path = '/content/drive/MyDrive/CSV PBO/indonesia - Sheet1.csv'
df = pd.read_csv(path)
df.describe()
```

```python
#menampilkan Data
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter',x='Life expectancy',y='percentage expenditure')

plt.title('Data Indonesia')
plt.xlabel('Life expectancy')
plt.ylabel('Percentage expenditure')
plt.xlim(64, 70)
plt.ylim(0, 9)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array(df['Life expectancy'])
y = np.array(df['percentage expenditure'])

print(f'X : {x}')
print(f'Y : {y}')
```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
x_vis = np.array([0, 70]).reshape(-1,1) #Pembuatan Garis Merah
y_vis = model.predict(x_vis) #Model machine learning
```

```python
plt.scatter(x,y)
plt.plot(x_vis, y_vis, '-r')
plt.title('Indeks')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(64, 70)
plt.ylim(0, 9)
plt.grid(True)
plt.show()
```

```python
print(f'intercept : {model.intercept_}')
print(f"slope : {model.coef_}")
```

```python
#Prediksi sample data
volume = np.array([6500000, 7250000, 7500000, 8750000, 14700000]).reshape(-1,1)
volume
```

```python
prediksi_indeks = model.predict(volume)
prediksi_indeks
```

```python
for penduduk, index in zip(volume, prediksi_indeks):
  print(f'Jumlah Penduduk : {penduduk} Prediksi Indeks : {index}')
```
